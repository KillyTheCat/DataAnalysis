```python
#function to find out no. of matches 2 teams played against each other
def numberofwins(teamA , teamB):
    m1 = match['team1']==teamA
    m2 = match['team1']==teamB
    m3 = match['team2']==teamA
    m4 = match['team2']==teamB
    
    return match[(m1&m4)|(m2&m3)]['winner'].value_counts().plot(kind = 'bar')
```

```python
def batsmanagainstbowlingteam(batsmanname):
    bdata = delivery[delivery['batsman']==batsmanname]
    return bdata.groupby('bowling_team').sum()['total_runs'].to_frame()
```

```python
batsmanagainstbowlingteam('A Chandila')
```

```python
#batsman runs against single bowlers
def batsmanagainstbowler(batsmanname):
    bdata = delivery[delivery['batsman']==batsmanname]
    return bdata.groupby('bowler').sum()['total_runs'].to_frame()
```

```python
batsmanagainstbowler('A Chandila')
```

```python
#how many runs a batsman made in a single season
def batsmanSeason(batname):
    new = delivery.merge(match,left_on='match_id',right_on='id')
    new = new[new['batsman']==batname]
    new.groupby('season').sum()['batsman_runs'].sort_values(ascending = False).plot(kind = 'bar')
```

```python
batsmanSeason('MS Dhoni')
```
